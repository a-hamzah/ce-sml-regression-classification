```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

```python
# Loading image
img = cv2.imread('library.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
# Converting BGR to RGB
real_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_img)
```

```python
# Resizing image
resized_img = cv2.resize(img, (100, 100), interpolation=cv2.INTER_NEAREST)
```

```python
plt.imshow(resized_img)
```

```python
# Rotating image
rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
# image rotation is important because algorithms are sensitive
# to angles, rotate one image to make more images (data augumentation)
```

```python
plt.imshow(rotated_img)
```

```python
# Creating a translation array
translation_arr = np.float32([[1, 0, -10], [0, 1, -10]]) # transformation matrix
```

```python
# Translating image on 1000 x 1000 plane
translated_img = cv2.warpAffine(img, translation_arr, (1000, 1000))
```

```python
plt.imshow(translated_img)
```

```python
# Concatenation
main_office = cv2.imread('main.jpg')
library = cv2.imread('library.jpg')
```

```python
# Loading two images
main_office = cv2.resize(main_office, (100, 100))
library = cv2.resize(library, (100, 100))
```

```python
# Merging two images
combined = cv2.hconcat([main_office, library])
```

```python
plt.imshow(combined)
```

```python
gray_img = cv2.cvtColor(real_img, cv2.COLOR_BGR2GRAY)
print(real_img.shape)
print(gray_img.shape)
```

```python
plt.imshow(gray_img, cmap='gray')
```

```python
edges = cv2.Canny(real_img, 100, 200) # colored images (cluttered output)
edges1 = cv2.Canny(gray_img, 100, 200) # grayscale image (not cluttered)
```

```python
plt.imshow(edges, cmap='gray')
```

```python
plt.imshow(edges1, cmap='gray')
```
